```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#data loading
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading data to a data frame

data_frame=pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# print first 5 rows

data_frame.head()
```

```python
#adding the target column

data_frame['diagnosis']=breast_cancer_dataset.target
```

```python
#number of rows and columns

data_frame.shape
```

```python
#getting info on the data(check for missing value)

data_frame.isnull().sum()
```

```python
#important statistical measures about the data

data_frame.describe()
```

```python
#checking target variable distribution

data_frame['diagnosis'].value_counts()
```

### 1-->benign 2-->malignant

```python
data_frame.groupby('diagnosis').mean()
```

```python
#seperating features and targets

X=data_frame.drop(columns='diagnosis', axis=1)  #dropping the diagnosis column from the x-axis
Y=data_frame['diagnosis']
```

```python
#visualize feature distributions by diagnosis
data_frame.groupby('diagnosis').hist(figsize = (20, 20))
```

```python
#splitting into training and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#model training using logistic regression

model=LogisticRegression()
```

```python
#model training using logistic regression

model.fit(X_train,Y_train)  #this learns the data and tries to establish a relationship between x_train and y_train to find best fit for classification
```

```python
#Accuracy Score
x_train_prediction=model.predict(X_train)
data_accuracy=accuracy_score(Y_train,x_train_prediction)
 
```

```python
print("accuracy on training data =",data_accuracy)
```

```python
#Accuracy Score on test data
x_test_prediction=model.predict(X_test)
testdata_accuracy=accuracy_score(Y_test,x_test_prediction)
```

```python
print("accuracy on test data =",testdata_accuracy)
```

The high accuracy shows that the model is doing good predictions...similarly we can also predict the diagnosis by imputing a value of our own
